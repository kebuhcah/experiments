```python
import pandas as pd
import re

def getGender(lemma):
    match = re.search('\{([a-z]+)\}', lemma)
    if match:
        return match[1]
    
def cleanLemma(lemma):
    return re.sub('\s*?(\{.*?\}|\(.*?\)|\[.*?\]|<.*?>)\s*?', '', lemma).strip()

Dict = pd.read_csv('dict_cc_de.txt', delimiter='\t', skiprows=8, header=None, na_filter=False)
Dict.columns = ['lemma', 'definition', 'pos', 'domain']
Dict['domain'] = Dict['domain'].str.strip()

Dict['gender'] = Dict['lemma'].apply(getGender)
Dict['original_lemma'] = Dict['lemma']
Dict['lemma'] = Dict['original_lemma'].apply(cleanLemma)

Dict['last_letter'] = Dict['lemma'].apply(lambda s: s[-1] if len(s) > 0 else '')

Dict.sample(20)
```

```python
Dict[(Dict['pos']=='noun')&(Dict['domain']=='')
     &(~Dict['lemma'].str.contains(' '))&(~Dict['definition'].str.contains(' '))
     &(Dict['last_letter']).apply(lambda c: 
                c.upper()!= c)].groupby(['last_letter', 
                    'gender'])['lemma'].nunique().unstack()[['m', 'f', 'n']].sort_values('f', ascending=False)
```
