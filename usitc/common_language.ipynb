```python
# more info here: https://www.usitc.gov/data/dicl.htm

import pandas as pd
import numpy as np
```

```python
Iso3 = pd.read_csv('iso3.tsv', sep='\t').drop(
    'M49 code', axis=1).rename(columns={'Country or Area': 'country1', 'ISO-alpha3 code': 'ISO3'}).set_index('ISO3')

Dicl = pd.read_csv('dicl.csv').set_index('ISO3').join(Iso3).reset_index().set_index('ISO3_2').join(
    Iso3.rename_axis('ISO3_2').rename(columns=lambda _:'country2')).reset_index()


Dicl
```
