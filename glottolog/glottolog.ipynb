```python
Families = pd.read_csv('Families.csv')
Families.describe()
```

```python
LangFamJoin = Languages[(~Languages.bookkeeping)&(Languages.family_pk>0)][['family_pk', 
                                      'name', 'level']].set_index('family_pk').join(Families[Families.pk>0][[
    'name', 'pk', 'level']].set_index('pk').rename(columns=lambda x: 'family_' +x)).drop('level', axis=1)

LangFamJoin.family_name.value_counts()
```
