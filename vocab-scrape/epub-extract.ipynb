```python
def processEntry(entry):
    string_contents = [c.strip() for c in entry.contents if not c.name]
    pos_contents = [c.text.strip() for c in entry.contents if c.name == 'i' and
                   not c.text.strip() in ['pansori', 'jing', 'hanbok', 'go', 'kimchi', 'soju', 'gut']]
    sup_contents = [c.text.strip() for c in entry.contents if c.name == 'sup']
    
    example = ''
    if entry.next_sibling.next_sibling.attrs.get('class')[0] == 'list101b':
        example = re.sub('^• ', '', entry.next_sibling.next_sibling.text).strip()
    
    result = {'number': int(entry.b.text),
            'pos': '/'.join(pos_contents),
            'sup': '/'.join(sup_contents),
            #'content_count': entry.contents,
            'korean': string_contents[0],
            'english': string_contents[-1],
            'example': example}
    
    if not result['pos'].endswith('.'):
        result['pos'] = result['pos'] + '.'
        result['english'] = re.sub(r'^\.', '', result['english']).strip()
        
    if result['sup']:
        result['korean'] = ' '.join(string_contents[:2])
    
    return result


FreqDict = DataFrame([processEntry(p) for p in bs.section.find_all('p', class_='list101a')])

FreqDict
```

```python
FreqDict.to_csv('data/korean.tsv', sep='\t',index=False)
```
