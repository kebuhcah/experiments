```python
import pandas as pd
import numpy as np
```

```python
with open('data/persian-freq-dict.txt') as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines if line.strip() != '' ]
len(lines)
```

```python
lines[867:]
```

```python
chunks = []
current_idx = 1
current_chunk = []

for i, l in enumerate(lines[871:]):
    if l == str(current_idx + 1):
        current_idx = current_idx + 1
        index = current_chunk[0]
        persian = current_chunk[1]
        definition = current_chunk[2]
        example = ' '.join(current_chunk[3:-1])
        chunks.append([index, persian, definition, example])
        current_chunk = [l]
    else:
        current_chunk.append(l)
     
```

```python
Entries = pd.DataFrame(chunks, columns=['idx', 'word', 'definition', 'example']).set_index('idx')
Entries.sample(10)
```
